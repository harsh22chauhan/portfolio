```python
x = data.iloc[:,[0,1]].values
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(data)
# Getting the cluster labels
labels = kmeans.predict(data)
# Centroid values
centroids = kmeans.cluster_centers_
```

```python
# Number of clusters
kmeans = KMeans(n_clusters=5)
# Fitting the input data
kmeans = kmeans.fit(data)
# Getting the cluster labels
labels = kmeans.predict(data)
# Centroid values
c = kmeans.cluster_centers_
print(c)
```

```python
plt.scatter(x[labels==0,0],x[labels == 0,1],s=50,c= 'red',label = 'cluster 1')
plt.scatter(x[labels==1,0],x[labels == 1,1],s=50,c= 'brown',label = 'cluster 2')
plt.scatter(x[labels==2,0],x[labels == 2,1],s=50,c= 'y',label = 'cluster 3')
plt.scatter(x[labels==3,0],x[labels == 3,1],s=50,c= 'blue',label = 'cluster 4')
plt.scatter(x[labels==4,0],x[labels == 4,1],s=50,c= 'green',label = 'cluster 5')
plt.scatter(c[:,0],c[:,1],s = 300,c = 'black',label = 'C',marker = "*")
plt.title("clusters of customer")
plt.legend()

plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i,init = 'k-means++',max_iter = 300,n_init = 10,random_state=0)
    km.fit(data)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
plt.title('elbow method')
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.show()
```
